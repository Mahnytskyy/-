```python
# Лабораторна робота 1. Магницький Іван Варіант 20
# Вступ до машинного навчання. Елементи лінійної алгебри.
# Завдання 1. Створіть одновимірний масив (вектор) з сумами кожного числа та його попереднього, починаючи з 1, та виведіть його значення.

# Імпорт бібліотеки NumPy
import numpy as np

# Створюємо масив із чисел від 1 до n
n = 10  # можна змінювати це значення для бажаної кількості елементів
array = np.arange(1, n+1)

# Створюємо масив із сум кожного числа і його попереднього
# Для цього використовуємо зсув масиву і додаємо
sums_array = array + np.roll(array, 1)

# Перший елемент масиву не має попереднього, тому його сумуємо з 0
sums_array[0] = array[0]

# Виводимо результат
print("Початковий масив:", array)
print("Масив сум попереднього і поточного елементів:", sums_array)
```

```python
# Завдання 6
# Створення двох матриць розміром 3x3 з випадкових цілих чисел від 1 до 20
matrix1 = np.random.randint(1, 21, (3, 3))
matrix2 = np.random.randint(1, 21, (3, 3))

# Виведення початкових матриць
print("Перша матриця:")
print(matrix1)
print("\nДруга матриця:")
print(matrix2)

# Матричний добуток (використовуємо np.dot або @)
product_matrix = np.dot(matrix1, matrix2)
# Можна також використовувати: product_matrix = matrix1 @ matrix2

# Виведення результату
print("\nДобуток двох матриць:")
print(product_matrix)
```

```python
# Завдання 7
# Створення матриці розміром 2x3 з випадкових цілих чисел від 1 до 4
matrix1 = np.random.randint(1, 5, (2, 3))

# Створення матриці розміром 3x3 з випадкових цілих чисел від 1 до 4
matrix2 = np.random.randint(1, 5, (3, 3))

# Виведення початкових матриць
print("Перша матриця (2x3):")
print(matrix1)
print("\nДруга матриця (3x3):")
print(matrix2)

# Множення матриць (розмір 2x3 на 3x3)
product_matrix = np.dot(matrix1, matrix2)
# Або можна використовувати оператор @: product_matrix = matrix1 @ matrix2

# Виведення результату
print("\nДобуток матриць:")
print(product_matrix)
```

```python
# Завдання 8
# Створення матриці розміром 3x3 з випадкових цілих чисел від 1 до 7
matrix = np.random.randint(1, 8, (3, 3))

# Виведення початкової матриці
print("Матриця 3x3:")
print(matrix)

# Перевірка, чи матриця є оборотною (детермінант не дорівнює 0)
determinant = np.linalg.det(matrix)
if determinant != 0:
    # Знаходження оберненої матриці
    inverse_matrix = np.linalg.inv(matrix)
    print("\nОбернена матриця:")
    print(inverse_matrix)
else:
    print("\nМатриця не є оборотною, оскільки її визначник дорівнює 0.")
```

```python
# Завдання 12
# Створення першої матриці розміром 3x3 з випадкових цілих чисел від 1 до 5
matrix1 = np.random.randint(1, 6, (3, 3))

# Створення другої матриці розміром 3x3 з випадкових цілих чисел від 1 до 5
matrix2 = np.random.randint(1, 6, (3, 3))

# Виведення початкових матриць
print("Перша матриця 3x3:")
print(matrix1)
print("\nДруга матриця 3x3:")
print(matrix2)

# Знаходження добутку елементів діагоналі другої матриці
diagonal_elements = np.diagonal(matrix2)  # Отримуємо елементи діагоналі
product = np.prod(diagonal_elements)       # Обчислюємо добуток елементів діагоналі

# Виведення результату
print("\nДобуток елементів діагоналі другої матриці:", product)
```

```python
# Завдання 15
# Створення матриці розміром 3x3 з випадкових дійсних чисел від -2 до 2
matrix = np.random.uniform(-2, 2, (3, 3))

# Виведення початкової матриці
print("Матриця 3x3:")
print(matrix)

# Знаходження кількості ненульових елементів в кожному рядку
non_zero_counts = np.count_nonzero(matrix, axis=1)

# Перетворення вектора в стовпчик
non_zero_counts_column = non_zero_counts.reshape(-1, 1)

# Виведення результату
print("\nКількість ненульових елементів в кожному рядку (вектор-стовпчик):")
print(non_zero_counts_column)
```
