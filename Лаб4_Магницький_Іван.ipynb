```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.svm import SVC

# Завантаження датасету Breast Cancer
data = load_breast_cancer()
X = data.data
y = data.target

# Створення DataFrame для зручності
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

# Виведення перших 5 рядків датасету
print("Перші 5 рядків датасету:")
df.head()

```

```python
# Розділення даних на тренувальну та тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Масштабування даних
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


```

```python
# 1. Логістична регресія
logreg = LogisticRegression(max_iter=1000)
logreg_params = {
    'C': [0.1, 1, 10],
    'penalty': ['l2']
}

logreg_grid = GridSearchCV(logreg, logreg_params, cv=5)
logreg_grid.fit(X_train, y_train)
logreg_best = logreg_grid.best_estimator_

# 2. Дерево рішень
tree = DecisionTreeClassifier(random_state=42)
tree_params = {
    'max_depth': [3, 5, 7],
    'min_samples_split': [2, 5, 10]
}

tree_grid = GridSearchCV(tree, tree_params, cv=5)
tree_grid.fit(X_train, y_train)
tree_best = tree_grid.best_estimator_

# 3. Метод опорних векторів (SVM)
svm = SVC()
svm_params = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

svm_grid = GridSearchCV(svm, svm_params, cv=5)
svm_grid.fit(X_train, y_train)
svm_best = svm_grid.best_estimator_

# Оцінка моделей
models = {'Logistic Regression': logreg_best, 'Decision Tree': tree_best, 'SVM': svm_best}

for model_name, model in models.items():
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{model_name} Accuracy: {accuracy}")
```

```python
# Завантаження датасету Breast Cancer
data = load_breast_cancer()
X = data.data
y = data.target

# Створення DataFrame для зручності
df = pd.DataFrame(X, columns=data.feature_names)
df['target'] = y

# Виведення перших 5 рядків датасету
print("Перші 5 рядків датасету:")
df.head()
```

```python
# Розділення даних на тренувальну та тестову вибірки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Масштабування даних
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Налаштування моделей
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'SVM': SVC(),
    'Random Forest': RandomForestClassifier()
}

# Гіперпараметри для налаштування
param_grids = {
    'Logistic Regression': {
        'C': [0.1, 1, 10, 100],
        'penalty': ['l2']  # Видалено 'l1' для lbfgs
    },
    'SVM': {
        'C': [0.1, 1, 10, 100],
        'kernel': ['linear', 'rbf']
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30]
    }
}
```

```python
# Функція для підбору гіперпараметрів і навчання моделі
def tune_and_train_model(model_name, model, param_grid):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
    grid_search.fit(X_train, y_train)
    best_params = grid_search.best_params_
    best_model = grid_search.best_estimator_
    return best_model, best_params

# Навчання і підбір гіперпараметрів для кожної моделі
best_models = {}
best_params = {}

for model_name, model in models.items():
    print(f"\nТонування моделі {model_name}...")
    best_model, params = tune_and_train_model(model_name, model, param_grids[model_name])
    best_models[model_name] = best_model
    best_params[model_name] = params

# Виведення найкращих параметрів
print("\nНайкращі параметри для кожної моделі:")
for model_name, params in best_params.items():
    print(f"{model_name}: {params}")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Оцінка моделей і прогноз
for model_name, model in best_models.items():
    print(f"\nОцінка моделі {model_name}:")

    # Прогнозування на тестовій вибірці
    y_pred = model.predict(X_test)

    # Виведення точності
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Точність моделі {model_name}: {accuracy:.4f}")

    # Звіт про класифікацію
    print("Звіт про класифікацію:")
    print(classification_report(y_test, y_pred))

    # Матриця помилок
    print("Матриця помилок:")
    print(confusion_matrix(y_test, y_pred))

# Вибір найкращої моделі на основі точності
best_model_name = max(best_models, key=lambda name: accuracy_score(y_test, best_models[name].predict(X_test)))
best_model = best_models[best_model_name]
print(f"\nНайкраща модель: {best_model_name}")

# Прогнозування за допомогою найкращої моделі
y_pred_best = best_model.predict(X_test)

# Виведення прогнозу
print("\nПрогнози на тестовій вибірці за допомогою найкращої моделі:")
print(y_pred_best)
```
